```python
# Criar um DataFrame fictício para simulação
data = {
    'Country Code': ['USA', 'CHN', 'IND', 'RUS', 'JPN', 'BRA', 'CAN', 'DEU', 'FRA', 'AUS'],
    'Region': ['North America', 'East Asia & Pacific', 'South Asia', 'Europe & Central Asia', 'East Asia & Pacific',
               'Latin America & Caribbean', 'North America', 'Europe & Central Asia', 'Europe & Central Asia', 'East Asia & Pacific'],
    'IncomeGroup': ['High income', 'Upper middle income', 'Lower middle income', 'Upper middle income', 'High income',
                    'Upper middle income', 'High income', 'High income', 'High income', 'High income']
}

df = pd.DataFrame(data)

# Combinar 'Region' e 'IncomeGroup' em uma única coluna para contagem
df['Region_IncomeGroup'] = df['Region'] + ' - ' + df['IncomeGroup']

# Contar a quantidade de cada combinação de 'Region' e 'IncomeGroup'
region_income_counts = df['Region_IncomeGroup'].value_counts()

# Plotar o gráfico de pizza
plt.figure(figsize=(12, 8))
region_income_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Distribuição de IncomeGroup por Region')
plt.ylabel('')  # Remover o rótulo do eixo y
plt.show()
```
